```python
projectsDF = pd.read_csv("../data/bioProjects.csv")
samplesDF = pd.read_csv("../data/bioSamples.csv")
```

```python
allGenotypeDf = samplesDF.dropna(axis=0, subset=['genotype'])
```

```python
allGenotypeDf['BioProjectID'] = allGenotypeDf['BioProjectID'].fillna(0)
allGenotypeDf['BioProjectID'] = allGenotypeDf['BioProjectID'].astype(np.int64)
```

```python
uniqueProjects = allGenotypeDf['BioProjectID'].unique()
```

```python
allGenotypeDf
```

```python
newProjectsDF = projectsDF[projectsDF['ID'].isin(uniqueProjects)]
```

```python
newProjectsDF
```

```python
allGenotypeDf.columns
```

```python
genotypeSamplesDF = allGenotypeDf[['sampleID', 'Title', 'BioProjectID', 'genotype']].copy()
```

```python
genotypeSamplesDF
```

```python
projectGenotypesDF = genotypeSamplesDF.groupby(['BioProjectID'])['genotype'].unique().reset_index()
```

```python
projectGenotypesDF.info()
```

```python
newProjectsDF.info()
```

```python
mergedDF = pd.merge(newProjectsDF, projectGenotypesDF, left_on='ID', right_on='BioProjectID')
mergedDF
```

```python
mergedDF.info()
```

```python
index_dict_list = []

# Iterate over the rows of the dataframe
for index, row in mergedDF.iterrows():
    genotype = row['genotype']
    description = row['Description']
    index_dict = {}
    
    # Iterate over the values in the genotype column
    for value in genotype:
        # Get the start and end indices of the value in the description text
        start_indices = [i for i in range(len(description)) if description.startswith(value, i)]
        end_indices = [i + len(value) for i in start_indices]
        
        # Store the start and end indices as a list of tuples in the dictionary
        index_dict[value] = list(zip(start_indices, end_indices))
    
    # Append the dictionary to the list
    index_dict_list.append(index_dict)

# Create a new column with the index dictionaries
mergedDF['index_dict'] = index_dict_list
```

```python
mergedDF
```

```python
# Create an empty list to store the successful rows
success_list = []

# Iterate over the rows of the dataframe
for index, row in mergedDF.iterrows():
    index_dict = row['index_dict']
    success = False
    
    # Iterate over the values in the index_dict dictionary
    for value in index_dict:
        # Check if the length of the list of tuples is greater than zero
        if len(index_dict[value]) > 0:
            success = True
            break
    
    # If at least one value has a non-empty list of tuples, add the row to the success list
    if success:
        success_list.append(row)

# Create a new dataframe with the successful rows
success_df = pd.DataFrame(success_list)
```

```python
success_df
```

```python
index = 14
row = success_df.loc[index]

# Get the description and genotype columns
description = row['Description']
genotype = row['genotype']
print("--------------")
print(row['index_dict'])
print("--------------")
# Iterate over the values in the genotype list
for value in genotype:
    # Find all occurrences of the value in the description
    start = 0
    while True:
        start = description.find(value, start)
        if start == -1:
            break
        end = start + len(value)
        # Highlight the specific occurrence of the value in the description
        description = description[:start] + '<u>' + description[start:end] + '</u>' + description[end:]
        start = end + len('<u></u>') - 1

# Print the highlighted description
display(HTML(description))
```

```python
success_df['Description Length'] = success_df['Description'].str.split().apply(len)
# Find the length of the longest description
longest_description_length = success_df['Description Length'].max()

# Find the length of the shortest description
shortest_description_length = success_df['Description Length'].min()

# Find the median length of the descriptions
median_description_length = success_df['Description Length'].median()

# Print the results
print("Longest Description Length:", longest_description_length)
print("Shortest Description Length:", shortest_description_length)
print("Median Description Length:", median_description_length)
```

```python
from sklearn.model_selection import train_test_split
(
    train_df,
    test_df
) = train_test_split(
    success_df, test_size=0.4, random_state=42
)
(
    val_df,
    test_df,
) = train_test_split(test_df, test_size=0.5, random_state=42)
```
